```python
import epilib as ep
import numpy as np
import matplotlib.pyplot as plt
```

```python
params = [0.1, 0.5, 0.6, 0.7, 0.8, 0.9, 0.99]
```

```python
hics = {}
diags = {}

for param in params:
    filename = "psa_" + str(param) + "/contacts.txt"
    print(filename)
    
    hics[param] = ep.get_contactmap(filename)
    diags[param] = ep.get_diagonal(hics[param])
```

```python
for param in params:
    plt.semilogy(diags[param])
```
